```python
def calc(a, b):
    return a + b, a - b, a * b, a / b

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
results = calc(x, y)
print(f"Sum: {results[0]}, Difference: {results[1]}, Product: {results[2]}, Quotient: {results[3]}")
```

```python
def main():
    print(f"20cm 피자 2개의 면적: {get_area(20)*2}")
    print(f"30cm 피자 1개의 면적: {get_area(30)}")

def get_area(radius):
    if radius > 0:
        area = 3.14*radius**2
    else:
        area = 0
    return area

main()
```

```python
def manu():
    print("1. 섭씨 -> 화씨")
    print("2. 화씨 -> 섭씨")
    print("3. 종료")
    choice = int(input("메뉴를 선택하세요: "))
    return choice

def c_to_f(c):
    return c * 9/5 + 32

def f_to_c(f):
    return (f - 32) * 5/9

while True:
    choice = manu()
    if choice == 1:
        c = float(input("섭씨 온도를 입력하세요: "))
        print(f"{c}°C는 {c_to_f(c)}°F입니다.")
    elif choice == 2:
        f = float(input("화씨 온도를 입력하세요: "))
        print(f"{f}°F는 {f_to_c(f)}°C입니다.")
    elif choice == 3:
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```

```python
pm = "+ - - "
ps = "|     "
plus = "+"
bar = "|"

def grid(x, y):
    for i in range(x):
        print(y*pm+plus)
        for j in range(y):
            print(y*ps+bar)
    print(y*pm+plus)

grid(4, 4)
```
