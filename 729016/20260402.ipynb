```python
n = int(input("정수 하나를 입력하세요: "))
num = 0
for i in range (1, n+1):
    num += i

print("1부터", n, "까지의 합은", num, "입니다.")
```

```python
for i in range(100):
    if i == 5:
        break
    print(i, end=" ")
```

```python
for i in range(20):
    if i % 2 == 0:
        continue
    print(i, end=" ")
```

```python
for i in range(2, 20):
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i, end=" ")
```
