```python
myList = [5, 10, 15, 20, 25]
print(f"변경전 : {myList}")
i = myList.index(10)
myList[i] = 100
print(f"변경후 : {myList}")  
```

```python
def remove10(myList):
    resultList = []
    for i in myList:
        if i != 10:
            resultList.append(i)
    return resultList

myList = [5, 10, 15, 20, 25]
print(f"변경전 : {myList}")
myList = remove10(myList)
print(f"변경후 : {myList}")
```
