```python
a = int(input('직각 삼각형의 밑변의 길이 입력'))
b = int(input('직각 삼각형의 높이의 길이 입력'))
c = (a**2 + b**2)**0.5
print(f"직각 삼각형의 빗변의 길이는 {c:.2f}입니다.")
```

```python
total = 0
americano = int(input('아메리카노 갯수 입력'))
cafelatte = int(input('카페라떼 갯수 입력'))
blacktea = int(input('블랙티 갯수 입력'))

total += americano * 4500
total += cafelatte * 5000
total += blacktea * 5900
print(f"총 주문 금액은 {total}원입니다.")
```

```python
country = input('국가 입력 (한국, 중국)')
price = int(input('물건의 가격 입력'))

if country == '한국':
    if price >= 20000:
        shipping_cost = 0
    else:
        shipping_cost = 3000
elif country == '중국':
    if price >= 100000:
        shipping_cost = 0
    else:
        shipping_cost = 8000

print(f"배송비는 {shipping_cost}원입니다.")
```

```python
score = int(input('점수 입력'))

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"학점은 {grade}입니다.")
```

```python
weight = float(input('몸무게 입력(킬로그램)'))
height = float(input('키 입력(미터)'))

bmi = weight / height ** 2
print(f"당신의 BMI는 {bmi:.2f}입니다.")

if bmi >= 20.0 and bmi < 25.0:
    print('정상입니다.')
elif bmi >= 25.0 and bmi < 30.0:
    print('과체중입니다.')
elif bmi >= 30.0:
    print('비만입니다.')
```

```python
import random

player1 = input('플레이어 1 이름 입력')
player2 = input('플레이어 2 이름 입력')

number1 = random.randint(1, 6)
number2 = random.randint(1, 6)

print(f"{player1}이(가) 뽑은 숫자는 {number1}입니다.")
print(f"{player2}이(가) 뽑은 숫자는 {number2}입니다.")

if number1 > number2:
    print(f"{player1}이(가) 이겼습니다.")
elif number1 < number2:
    print(f"{player2}이(가) 이겼습니다.")
else:
    print("비겼습니다.")
```
