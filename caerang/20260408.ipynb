```python
data = [
    ['037730', '3R', '1510'],
    ['036360', '3SOFT', '1790'],
    ['005760', 'ACTS', '1185']
]

columns = ['종목코드', '종목명', '현재가'] 
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
df.set_index('종목코드', inplace=True)
df
```

```python
df['종목명']
```

```python
df.iloc[0]
```

```python
print(df.iloc[0,0])
print(df.iloc[1,0])
print(df.iloc[2,0])
```

```python
print(df.loc['037730', '종목명'])
print(df.loc['036360', '종목명'])
print(df.loc['005760', '종목명'])
```

```python
data = [
    ["005930", "삼성전자", 66500, 66500 * 1.15],
    ["035720", "카카오", 80600, 80600 * 1.15],
    ["005380", "현대차", 185000, 185000 * 1.15]
]

df = pd.DataFrame(data=data, columns=['종목코드', '종목명', '현재가', '목표가'])
df
```
