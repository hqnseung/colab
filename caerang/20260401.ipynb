```python
s1 = set([1,2,3])
s2 = set([3,4,5])

print(s1 & s2)
print(s1 | s2)
print(s1 - s2)
```

```python
s3 = set([1,2,3])
s3.add(4)
print(s3)
```

```python
s3 = set([1,2,3])

s3.update([5,6])
print(s3)

s3.discard(1)
print(s3)

s3.discard(10)
print(s3)

s3.clear()
print(s3)
```

```python
scores = {
    '국어': 85,
    '영어': 90,
    '수학': 95
}

print(f"국어: {scores['국어']}\n영어: {scores['영어']}\n수학: {scores['수학']}")

for subject, score in scores.items():
    print(f"{subject}: {score}")
```
